```python
env = dbutils.widgets.get("env")
schema = dbutils.widgets.get("schema")
volume = dbutils.widgets.get("volume")
```

```python
config = get_config(env = f"{env}", 
                    path = "../utils/config.json")
catalog = config.get("catalog")                    
```

```python
from pyspark.sql.functions import *
```

```python
class schema_volume_creation:
    def __init__(self, catalog, schema, volume):
        self.catalog = catalog
        self.schema = schema
        self.volume = volume
    def create_schema(self):
        spark.sql(f"""
            DROP SCHEMA IF EXISTS {self.catalog}.{self.schema} CASCADE
        """)
        spark.sql(f"""
            CREATE SCHEMA IF NOT EXISTS {self.catalog}.{self.schema}
        """)
        print(f"{self.schema} schema is dropped and recreated...")

    def create_volume(self):
        spark.sql(f"""
            DROP VOLUME IF EXISTS {self.catalog}.{self.schema}.{self.volume}
        """)
        spark.sql(f"""
            CREATE VOLUME IF NOT EXISTS {self.catalog}.{self.schema}.{self.volume}
        """)
        print(f"{self.volume} volume is dropped and recreated...")
```

```python
obj = schema_volume_creation(catalog, schema, volume)
obj.create_schema()
obj.create_volume()
```
