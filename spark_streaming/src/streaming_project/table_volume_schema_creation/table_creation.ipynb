```python
import os
import json
```

```python
dbutils.widgets.text("catalog", "lakehouse")
dbutils.widgets.text("schema", "bronze")
dbutils.widgets.text("table", "bronze_invoice_details")
```

```python
catalog = dbutils.widgets.get("catalog")
schema = dbutils.widgets.get("schema")
table = dbutils.widgets.get("table")
```

```python
with open("../utils/table_creation_details.json") as f:
    table_creation_details = json.load(f)

required_details = [detail for detail in table_creation_details if detail.get("schema") == schema][0]

print(required_details)
path = required_details.get("path")
```

```python
arguments = {
    "catalog": catalog,
    "schema": schema,
    "table": table,
}

with open(f"{path}/{table}.sql") as file:
    sql_file = file.read()
    ##pass arguments
    sql_file = sql_file.format(**arguments)

##drop table if already exists and create fresh table
if spark.catalog.tableExists(f"{catalog}.{schema}.{table}"):
    spark.sql(f"DROP TABLE {catalog}.{schema}.{table}")
spark.sql(sql_file)
print(f"Table {table} created successfully")
```
